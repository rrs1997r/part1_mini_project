```python
import yfinance as yf
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
# Retrieve stock data
tickerSymbol = 'GOOGL'
tickerData = yf.Ticker(tickerSymbol)
tickerDf = tickerData.history(period='1d', start='2010-5-31', end='2020-5-31')
```

```python
#matplotlib

plt.figure(figsize=(10, 6))
plt.plot(tickerDf.index, tickerDf['Close'], label='Closing Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Closing Price')
plt.legend()
plt.show()
```

```python
# Plotly graph

fig = go.Figure()
fig.add_trace(go.Scatter(x=tickerDf.index, y=tickerDf['Volume'], name='Volume'))
fig.update_layout(title='Stock Volume', xaxis_title='Date', yaxis_title='Volume')
fig.show()
```
